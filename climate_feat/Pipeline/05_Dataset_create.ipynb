```python
import pandas as pd

feat_daily = pd.read_parquet(INPUT)
feat_daily["date"] = pd.to_datetime(feat_daily["date"])
feat_daily["county"] = feat_daily["county"].astype(str).str.strip()


assert feat_daily.duplicated(["date","county"]).sum() == 0
```

```python
calendar['date'] = pd.to_datetime(calendar['date'])
```

```python
def safe_merge(left, right, on, how="left", name="right"):
    before = len(left)
    out = left.merge(right, on=on, how=how)
    after = len(out)
    if after != before:
        raise ValueError(f"Row count changed after merging {name}: {before} -> {after}. Likely a many-to-many join on {on}.")
    return out
```

```python
feat_daily_add = safe_merge(feat_daily,calendar,'date')
feat_daily_add.head()
```

```python
elec_dmd["timestamp"] = pd.to_datetime(elec_dmd["timestamp"], utc=True)
elec_dmd
```

```python
elec_dmd["timestamp"].dtype
```

```python
# Unpivot the DataFrame using pd.melt()
elec_melt = pd.melt(elec_dmd,
                  id_vars=['timestamp'],
                  var_name='county',
                  value_name='load_mw')

print(elec_melt)
```

```python
elec_peak_daily_county = elec_melt.rename(columns={"timestamp": "date"})
elec_peak_daily_county
```

```python
# Clean county strings
elec_peak_daily_county["county"] = elec_peak_daily_county["county"].astype(str).str.strip()

# Force date to midnight (drop time)
d = pd.to_datetime(elec_peak_daily_county["date"])
elec_peak_daily_county["date"] = d.dt.floor("D").dt.tz_localize(None)   # sets to midnight
```

```python
elec_peak_daily_county["date"].dtype
```

```python
feat_daily_add['date'].dtype
```

```python
elec_peak_daily_county = (
    elec_peak_daily_county
    .groupby(["date", "county"], as_index=False, observed=True)["load_mw"]
    .max()
)
elec_peak_daily_county
```

```python
feat_daily_add2 = feat_daily_add.merge(
    elec_peak_daily_county,
    on=["date","county"],
    how="left"
)
feat_daily_add2.shape
```
